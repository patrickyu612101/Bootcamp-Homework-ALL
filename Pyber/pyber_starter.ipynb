```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cdata=pd.read_csv(city_data_to_load)
rdata=pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
alldata=rdata.merge(cdata,how="left")


# Display the data table for preview
alldata.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
    #Urban data
Udata=alldata.loc[alldata["type"]=="Urban"]
ug=Udata.groupby("city")
xU=ug.count()["driver_count"]
yU=ug.mean()["fare"].round(2)


    #Suburban data
Sdata=alldata.loc[alldata["type"]=="Suburban"]
sg=Sdata.groupby("city")
xS=sg.count()["driver_count"]
yS=sg.mean()["fare"].round(2)


    #Rural data
Rdata=alldata.loc[alldata["type"]=="Rural"]
rg=Rdata.groupby("city")
xR=rg.count()["driver_count"]
yR=rg.mean()["fare"].round(2)





# Build the scatter plots for each city types
# urban, =
plt.scatter(xU,yU,marker="o",facecolor="lightcoral",edgecolors="black",s=xU*5, alpha=0.5, label="Urban")
# suburban, =
plt.scatter(xS,yS,marker="o",facecolor="lightblue",edgecolors="black",s=xS*5,alpha=0.5, label="Suburban")
# rural, = 
plt.scatter(xR,yR,marker="o",facecolor="gold",edgecolors="black",s=xR*5,alpha=0.5, label ="Rural")


# Incorporate the other graph properties

# Create a legend
plt.legend()
# Incorporate a text label regarding circle size
plt.xlabel("Total numberof rides: per city")
plt.ylabel("Average fare")
plt.title("Pyber Ride data")
# Save Figure
plt.savefig("homework5_part1.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
    #get names
types=["Rural","Suburban","Urban"]
sumfares= alldata.groupby("type")
sumfares= sumfares.sum()
sumfares=sumfares["fare"]
    #get values of type
value=sumfares
explode = (0,0,0.1)
# Build Pie Chart
plt.pie(value, labels=types,explode=explode,autopct="%1.1f%%", shadow=True,startangle=140)
plt.title("% of total fare per city")
# Save Figure
plt.savefig("Total Fare pie chart")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides= alldata.groupby("type")
rides= rides.count()
rides=rides["driver_count"]
value=rides

# Build Pie Chart
plt.pie(value, labels=types,explode=explode,autopct="%1.1f%%", shadow=True,startangle=140)
plt.title("% of total Ride per city")
# Save Figure
plt.savefig("Total rides pie chart")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

driver= alldata.groupby("type")
driver= driver.sum()

driver=driver["driver_count"]
value=driver
print(value)
# Build Pie Charts
plt.pie(value, labels=types,autopct="%1.1f%%",explode=explode, shadow=True,startangle=140)
plt.title("% of total Driver per city")
# Save Figure
plt.savefig("Total driver pie chart")
```
